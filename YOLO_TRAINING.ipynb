```python
import os

# Download dataset
dataset_url = "https://universe.roboflow.com/ds/FNJkqQJXnB?key=2wcCHUQe0K"
os.system(f"curl -L {dataset_url} > roboflow.zip")

# Extract dataset
os.system("unzip roboflow.zip -d /content/dataset")
```

```python
from ultralytics import YOLO

# Load the pre-trained model (nano version is lightweight)
model = YOLO('yolov8n.pt')
```

```python
model.train(data="/content/dataset/data.yaml", epochs=50, imgsz=640, batch=16)
```

```python
results = model.val()
print(results)
```

```python
results = model.predict(source='/content/sample.jpg', save=True)

```
